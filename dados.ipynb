6 - Verificando quantidade e quais cidades existem os cursos procurados.

```python
curso_procurado = "ENGENHARIA DA COMPUTAÇÃO"

cidades_do_curso = tabela[tabela['NOME_CURSO'] == curso_procurado]['MUNICIPIO'].unique()

quantidade_de_cidades = len(cidades_do_curso)

print("Quantidade de cidades onde o curso está disponível:", quantidade_de_cidades)
print(cidades_do_curso)
```

```python
curso_procurado = "ENGENHARIA DE SOFTWARE"

cidades_do_curso = tabela[tabela['NOME_CURSO'] == curso_procurado]['MUNICIPIO'].unique()

quantidade_de_cidades = len(cidades_do_curso)

print("Quantidade de cidades onde o curso está disponível:", quantidade_de_cidades)

print(cidades_do_curso)
```

```python
curso_procurado = "ANÁLISE E DESENVOLVIMENTO DE SISTEMAS"

cidades_do_curso = tabela[tabela['NOME_CURSO'] == curso_procurado]['MUNICIPIO'].unique()

quantidade_de_cidades = len(cidades_do_curso)

print("Quantidade de cidades onde o curso está disponível:", quantidade_de_cidades)

print(cidades_do_curso)
```

7 - Verificando a modalidade dos cursos e quantidades por cidade.

```python
contagem_cursos_por_cidade = tabela.groupby(['MUNICIPIO', 'MODALIDADE'])['CODIGO_CURSO'].count().reset_index()

display(contagem_cursos_por_cidade)
```

8 - Verificando quantidade de cursos oferecidos por Instituição e total de Instituições.

```python
contagem_instituicoes = tabela['NOME_IES'].value_counts().reset_index()

contagem_instituicoes.columns = ['Nome da Instituição', 'Quantidade de Cursos']

total_instituicoes = tabela['NOME_IES'].nunique()

print("Total de Instituições:", total_instituicoes)

display(contagem_instituicoes)
```

9 - Comparativo entre Categoria administrativa com os tipos de organização academica.

```python
tabela_comparativa = pd.crosstab(tabela['ORGANIZACAO_ACADEMICA'], tabela['CATEGORIA_ADMINISTRATIVA'])

display(tabela_comparativa)
```
