9.1 - Retirando todos os itens repetidos para extração do gráfico.

```python
instituicoes_em_atividade = tabela[(tabela.SITUACAO_CURSO != 'Em extinção') & (tabela.SITUACAO_CURSO != 'Extinto')].drop_duplicates(subset=['NOME_IES'])

instituicoes_em_atividade

adm_instituicoes = instituicoes_em_atividade['CATEGORIA_ADMINISTRATIVA'].value_counts()

adm_name = instituicoes_em_atividade['CATEGORIA_ADMINISTRATIVA'].drop_duplicates()

tabela_adm = {}
nova_tabela_adm = pd.DataFrame(tabela_adm)

for x in range(len(adm_name)):
    adm = adm_name.iloc[x]
    qto = adm_instituicoes.iloc[x]

    nova_linha = pd.DataFrame({ 'Administrador': [adm], 'Quantidade': [qto] }, index=[x])
    nova_tabela_adm = pd.concat([nova_tabela_adm, nova_linha])

display(nova_tabela_adm)
```

```python
import numpy as np 
import matplotlib.pyplot as plt   
  
adm = nova_tabela_adm['Administrador']
qto = nova_tabela_adm['Quantidade']
  
explode = (0.1, 0.0, 0.1, 0.2, 0.3, 0.4) 
colors = ( 'blue', 'Red', 'purple', 
          'grey', 'lightblue', 'beige') 
wp = { 'linewidth' : 1, 'edgecolor' : 'black' } 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return '{:.2f}%'.format(pct, absolute) 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(qto,  
                                  autopct = lambda pct: func(pct, qto), 
                                  explode = explode,  
                                  labels = adm, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 50, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
ax.legend(wedges, adm, 
          title ='Tipos de administração', 
          loc ='center left', 
          bbox_to_anchor =(0.9, 0.1, 1.0, 0.6)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title('Comparativo tipos de administração de instituições Brasil - 2022') 
plt.savefig("grafico_01.png")
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Substitua 'nova_tabela_estados' pelos seus dados reais
# Exemplo de dados:
nova_tabela_estados = pd.DataFrame({
    'Estados': ['SP', 'MG', 'PR', 'BA', 'RJ', 'GO', 'PE', 'RS', 'CE', 'SC', 'PA', 'MT', 'ES', 'DF', 'MA', 'PI', 'PB', 'MS', 'AL', 'AM', 'RN', 'RO', 'TO', 'SE', 'AP', 'AC', 'RR', 'ZZ'],
    'N° Universidades': [671, 340, 190, 173, 154, 138, 134, 132, 120, 118, 87, 83, 82, 74, 73, 56, 52, 43, 39, 39, 38, 34, 27, 26, 16, 14, 9, 3]
})

# Calcular a porcentagem de universidades em cada estado
total_universidades = nova_tabela_estados['N° Universidades'].sum()
porcentagens = (nova_tabela_estados['N° Universidades'] / total_universidades) * 100

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
bars = plt.bar(nova_tabela_estados['Estados'], porcentagens, color='skyblue')
plt.xlabel('Estados')
plt.ylabel('Porcentagem de Universidades')
plt.title('Porcentagem de Universidades por Estado')
plt.xticks(rotation=45)

# Adicionar as porcentagens nas barras
for bar, porcentagem in zip(bars, porcentagens):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 1, f'{porcentagem:.2f}%', ha='center', color='black', fontsize=8)

# Exibir o gráfico
plt.savefig("grafico_03.png")
plt.tight_layout()
plt.show()
```

```python
from fpdf import FPDF

class PDF(FPDF):

    def header(self):
        # Define a cor de fundo como cinza (RGB: 192, 192, 192)
        self.set_fill_color(132, 245, 226)
        
        # Desenha um retângulo que preenche toda a página (210x297 mm para tamanho A4)
        self.rect(0, 0, 210, 297, 'F')

    def doc_title(self, label):
        self.set_font("helvetica", "B",16)
        self.cell(0, 10, label, 0, 1, "C")
    
    def doc_text(self, text):
        self.set_font("helvetica", "",12)
        self.multi_cell(0, 7, text)
        self.ln()
    
    def doc_image(self, img, x, y, w, h):
        self.image(img, x, y, w, h)

pdf = PDF()
pdf.add_page()

pdf.doc_title("Relatório Educação no Brasil.")

pdf.ln(80)

pdf.doc_title("Projeto desenvolvido por: Lucas Rodrigues Policapo")

pdf.ln(50)

pdf.doc_text("Este relatório tem como objetivo apresentar os dados obtidos a partir da análise dos dados do Censo da Educação Superior de 2020, disponibilizados pelo INEP (Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira).")

pdf.ln(60)

pdf.doc_title("Itanhaém, 17 de setembro de 2023")

pdf.add_page()

pdf.doc_title("Análise aprofundada sobre a educação atual brasileira.")

pdf.doc_image("inicio.png", 8, 20, (640/3.3), (480/6))

pdf.ln(80)

pdf.doc_text("\nA busca por uma educação superior representa um marco significativo na trajetória de qualquer indivíduo. É uma etapa crucial que exige uma tomada de decisão informada. Compreender profundamente as instituições de ensino superior no Brasil desempenha um papel crucial nesse processo de escolha. Portanto, neste artigo, buscamos fornecer um panorama abrangente e informativo sobre o cenário educacional do ensino superior no país.\n\nÉ fundamental esclarecer que nossa intenção aqui não é avaliar ou classificar as instituições de ensino, mas sim apresentar uma análise baseada em dados quantitativos. Iremos explorar diversos aspectos, desde a diferenciação entre instituições públicas e privadas, até a investigação dos diferentes modelos de ensino oferecidos no Brasil. Além disso, também examinaremos a distribuição geográfica das instituições de ensino superior por região, identificando tendências e peculiaridades que podem influenciar na decisão de quem busca prosseguir com sua formação acadêmica.\n\nNossa análise se propõe a ser uma fonte confiável e abrangente de informações para aqueles que estão considerando cursar o ensino superior. Entendemos que a escolha da instituição de ensino e do curso adequado é um passo fundamental para o sucesso acadêmico e profissional, e esperamos que os dados apresentados aqui contribuam para uma tomada de decisão mais embasada e consciente por parte dos estudantes e suas famílias. Ao final deste artigo, você estará mais bem preparado para iniciar sua jornada educacional com confiança e determinação.")

pdf.ln(10)

pdf.doc_title("Quantificação de Instituições por região")

pdf.add_page()



pdf.doc_image("grafico_02.png", 25, 10, (640/4), (480/4))

pdf.ln(130)

pdf.doc_text("A análise da distribuição das instituições de ensino superior no Brasil por região oferece uma visão interessante da paisagem educacional do país. Os dados indicam que a região Sudeste desponta como líder, abrigando um impressionante total de 1.469 instituições. Essa predominância é explicada, em parte, pela alta densidade populacional e pelo desenvolvimento econômico que caracterizam essa parte do país.\n\nO Nordeste, com 853 instituições, se posiciona de forma significativa, destacando o compromisso com a educação em uma região que historicamente enfrentou desafios socioeconômicos. A presença de tantas instituições na região Sul, que abrange estados como Paraná, Santa Catarina e Rio Grande do Sul, com 1.028 no total, reflete a qualidade do ensino e o dinamismo econômico dessa área.\n\nA região Centro-Oeste, com 723 instituições, desempenha um papel fundamental, principalmente devido à presença do Distrito Federal e seu caráter político e econômico central. A região Norte, com 598 instituições, enfrenta desafios logísticos, mas demonstra um compromisso em expandir o acesso à educação superior em uma área geograficamente vasta.\n\nCuriosamente, a categoria IGNORADO/EXTERIOR também figura na lista, com 191 instituições. Essa categoria abrange instituições cuja localização não foi especificada ou aquelas localizadas no exterior, mas conectadas ao sistema de ensino superior brasileiro.\n\nEsses números, embora representem apenas parte do panorama educacional, destacam a diversidade geográfica do sistema de ensino superior do Brasil. Eles também ressaltam a importância de considerar a localização e a disponibilidade de instituições ao tomar decisões educacionais, garantindo um acesso equitativo e igualitário à educação de qualidade em todo o país.")


pdf.doc_title("Analisando diretamente por estado")
pdf.doc_image("grafico_03.png", 25, 80, (640/4), (480/4))

pdf.ln(130)

pdf.doc_text("Estado de São Paulo (SP): Com 671 universidades, São Paulo lidera de forma destacada, contribuindo significativamente para a alta quantidade de instituições no Sudeste.\n\nEstado de Minas Gerais (MG): Em segundo lugar, Minas Gerais conta com 340 universidades, consolidando sua importância no cenário educacional do país.\n\nEstado do Paraná (PR): O Paraná, com 190 universidades, reforça a presença de instituições de ensino superior no Sul do Brasil.\n\nEstado da Bahia (BA): A Bahia, com 173 universidades, destaca-se na região Nordeste como um polo educacional importante.\n\nEstado do Rio de Janeiro (RJ): O Rio de Janeiro, com 154 universidades, também mantém uma presença considerável na região Sudeste.\n\nEsses estados com maiores números de universidades demonstram um compromisso sólido com o ensino superior, oferecendo uma ampla gama de opções para estudantes em busca de formação acadêmica.\n\nPor outro lado, estados como Roraima (RR), Acre (AC), e Amapá (AP) apresentam um número menor de universidades, com 9, 14 e 16 respectivamente, refletindo os desafios de acesso à educação superior em regiões mais remotas e menos densamente povoadas do Brasil.\n\nConsiderando esses dados em conjunto com os números das regiões, fica evidente que a distribuição das universidades no Brasil é influenciada pela demografia, economia e políticas educacionais de cada estado e região. Essas informações são fundamentais para estudantes que estão ponderando suas opções de ensino superior, pois ajudam a compreender melhor a disponibilidade de instituições em diferentes partes do país, além de ressaltar a importância de políticas educacionais que promovam a equidade e o acesso igualitário à educação em todo o território nacional.")

pdf.doc_title("Avaliando os tipos de instituições no Brasil")
pdf.doc_image("grafico_01.png", 25, 175, (640/4), (480/4))

pdf.add_page()


pdf.ln(20)


pdf.doc_text("1. **Privada com fins lucrativos:** Com um total de 1.720 instituições, as universidades privadas com fins lucrativos são as mais comuns no Brasil. Isso sugere a presença de um setor educacional altamente comercial, onde as instituições têm o objetivo de obter lucro.\n\n2. **Privada sem fins lucrativos:** Com 919 instituições, as universidades privadas sem fins lucrativos também desempenham um papel significativo no sistema educacional do país. Essas instituições têm um caráter mais voltado para a missão educacional e, em muitos casos, podem ser ligadas a organizações religiosas ou filantrópicas.\n\n3. **Pública Municipal:** As instituições públicas municipais, com 138 universidades, representam o compromisso das administrações locais em fornecer educação superior em nível municipal.\n\n4. **Pública Federal:** As universidades públicas federais, com 121 instituições, têm uma presença importante no sistema educacional do Brasil. Elas são financiadas pelo governo federal e muitas vezes são reconhecidas pela qualidade de ensino.\n\n5. **Pública Estadual:** Com 42 instituições, as universidades públicas estaduais também desempenham um papel relevante na oferta de ensino superior. Elas são financiadas pelos governos estaduais e, assim como as federais, podem ser reconhecidas por sua qualidade acadêmica.\n\n6. **Especial:** As instituições de ensino superior classificadas como Especial, com um total de 25, podem representar instituições com características únicas, como escolas técnicas, institutos específicos ou outras instituições de ensino especializado.\n\nEssa análise revela a diversidade e a complexidade do sistema de ensino superior no Brasil, que inclui instituições públicas e privadas com diferentes missões e objetivos. Essa variedade oferece aos estudantes uma ampla gama de escolhas ao buscar sua formação acadêmica e ressalta a importância de considerar esses fatores ao tomar decisões educacionais. Além disso, esses dados refletem o compromisso do país em fornecer educação de qualidade em diversas modalidades e em diferentes regiões.\n\n É interessante notar que as universidades privadas com fins lucrativos representam a maioria das instituições de ensino superior no Brasil, indicando uma forte presença do setor privado na educação superior. As instituições públicas, incluindo as federais, estaduais e municipais, também desempenham um papel importante na oferta de ensino superior, proporcionando opções acessíveis e de alta qualidade para os estudantes.")

pdf.doc_title("Comparação entre as modalidades no Brasil")
pdf.doc_image("grafico_04.png", 25, 50, (640/4), (480/4))

pdf.ln(150)

pdf.doc_text("A preferência das pessoas pela modalidade de Educação a Distância (EaD) em relação ao ensino presencial tem sido um fenômeno crescente nas últimas décadas. Essa mudança de preferência pode ser atribuída a uma série de fatores, cada um dos quais desempenha um papel significativo na decisão dos estudantes.\n\nFlexibilidade de Horários: A Educação a Distância oferece flexibilidade de horários que o ensino presencial muitas vezes não pode proporcionar. Os estudantes podem acessar o conteúdo do curso e realizar suas atividades de aprendizado quando for mais conveniente para eles, permitindo que conciliem os estudos com outras responsabilidades, como trabalho e família.\n\nAcessibilidade Geográfica: O EaD elimina barreiras geográficas, permitindo que estudantes de qualquer lugar do país (ou até mesmo do mundo) tenham acesso a cursos de instituições de ensino que podem estar distantes de suas residências.\n\nCusto mais Baixo: Em muitos casos, os cursos de EaD são mais acessíveis financeiramente do que os cursos presenciais. Isso ocorre porque as instituições podem economizar em infraestrutura física e, consequentemente, repassar essas economias aos estudantes na forma de mensalidades mais baixas.\n\nVariedade de Opções de Cursos: A Educação a Distância oferece uma ampla variedade de cursos, muitos dos quais podem não estar disponíveis em instituições presenciais próximas. Isso permite que os estudantes escolham cursos que atendam melhor às suas metas de carreira e interesses pessoais.\n\nTecnologia Avançada: Com o avanço da tecnologia, os cursos de EaD podem oferecer experiências de aprendizado interativo e de alta qualidade. Plataformas educacionais online, videoconferências e recursos de simulação estão se tornando cada vez mais sofisticados, proporcionando uma experiência de aprendizado envolvente.\n\nPossibilidade de Trabalhar enquanto Estuda: Muitos estudantes optam pelo EaD porque podem manter empregos em tempo integral enquanto obtêm um diploma. Isso ajuda a reduzir o impacto financeiro de frequentar a faculdade.\n\nAutodisciplina e Autonomia: A Educação a Distância requer um alto grau de autodisciplina e responsabilidade, o que pode ser uma qualidade atraente para estudantes que preferem aprender no seu próprio ritmo e assumir o controle de seu próprio aprendizado.\n\nAdaptação às Novas Realidades: A pandemia da COVID-19 acelerou ainda mais a adoção do EaD, tornando-o uma opção mais viável para muitos estudantes que, por motivos de saúde ou restrições de deslocamento, não podiam frequentar aulas presenciais.\n\nÉ importante notar que a preferência entre EaD e ensino presencial varia de acordo com as necessidades e preferências individuais de cada estudante. Algumas pessoas ainda preferem o contato direto com professores e colegas em um ambiente presencial, enquanto outras valorizam a flexibilidade e a conveniência oferecidas pelo EaD. A escolha entre essas modalidades depende dos objetivos de aprendizado e das circunstâncias pessoais de cada um.\n\nEm suma, a análise dos dados apresentados revela a complexidade do cenário educacional brasileiro, refletindo a diversidade de escolhas disponíveis para os estudantes. A preferência pela modalidade de Educação a Distância (EaD) em relação ao ensino presencial tem crescido devido à flexibilidade, acessibilidade geográfica, custos mais baixos, variedade de cursos, avanços tecnológicos e a capacidade de equilibrar estudos com outras responsabilidades. No entanto, é fundamental reconhecer que a decisão entre EaD e presencial é pessoal, dependendo das necessidades, metas e circunstâncias individuais de cada estudante. Essa variedade de opções é um reflexo do compromisso contínuo do Brasil em proporcionar acesso à educação de qualidade em diferentes formatos e regiões.")

pdf.doc_image("final.png", 18, 60, (700/4), (300/4))

pdf.ln(90)

pdf.doc_text("Em análise abrangente dos dados apresentados, é evidente que o cenário do ensino superior no Brasil é multifacetado e diversificado. As instituições de ensino superior estão distribuídas de maneira desigual por região, com destaque para a predominância no Sudeste. Além disso, os diferentes tipos de instituições, incluindo públicas e privadas, com e sem fins lucrativos, demonstram a complexidade do sistema educacional brasileiro.\n\nA preferência por modalidades de ensino, seja EaD ou presencial, reflete as mudanças nas expectativas dos estudantes. A flexibilidade do EaD tem atraído muitos devido à sua capacidade de conciliar estudos com outras responsabilidades. No entanto, a decisão entre essas modalidades continua sendo uma escolha pessoal, influenciada por diversos fatores, incluindo objetivos de carreira, necessidades individuais e circunstâncias pessoais.\n\nO Brasil está empenhado em oferecer educação superior de qualidade em todo o país, e a variedade de opções disponíveis reflete esse compromisso. Independentemente da modalidade escolhida, é importante reconhecer que a educação superior desempenha um papel fundamental na formação de indivíduos e no desenvolvimento do país como um todo. A diversidade de opções educacionais é uma força que permite que os estudantes encontrem o melhor caminho para atingir suas metas acadêmicas e profissionais")

pdf.doc_title("Referências Bibliograficas")

pdf.ln(20)

pdf.doc_text("1 - Lei de Diretrizes e Bases da Educação Nacional (LDB):\nLei nº 9.394/96. Disponível em: [https://legislacao.presidencia.gov.br/ficha?/legisla/legislacao.nsf/Viw_Identificacao/lei%209.394-1996&OpenDocument].\n\n2 - Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (INEP):\nINEP. Disponível em: [https://www.gov.br/inep/pt-br/areas-de-atuacao/pesquisas-estatisticas-e-indicadores/censo-da-educacao-superior].\n\n3 - Ministério da Educação (MEC):\nMinistério da Educação (MEC). Disponível em: [http://portal.mec.gov.br/component/tags/tag/ensino-superior].\n\n4 - Associação Brasileira de Mantenedoras de Ensino Superior (ABMES):\nABMES.  Disponível em: [https://abmes.org.br/abmes-pesquisas/categoria/345/pesquisas].")


pdf.output("AnaliseGeral.pdf")
```
