6 - Verificando quantidade e quais cidades existem os cursos procurados.

```python
curso_procurado = "ENGENHARIA DA COMPUTAÇÃO"

cidades_do_curso = tabela[tabela['NOME_CURSO'] == curso_procurado]['MUNICIPIO'].unique()

quantidade_de_cidades = len(cidades_do_curso)

print("Quantidade de cidades onde o curso está disponível:", quantidade_de_cidades)
print(cidades_do_curso)
```

6.1 - Comparação entre quantidade de municipios que oferecem cada curso.

```python
comparacao_cursos = tabela.groupby('NOME_CURSO')['MUNICIPIO'].nunique().reset_index()

comparacao_cursos.columns = ['Nome do Curso', 'Quantidade de Municípios']

print(comparacao_cursos)
```

7 - Verificando a modalidade dos cursos e quantidades por cidade.

```python
contagem_cursos_por_cidade = tabela.groupby(['MUNICIPIO', 'MODALIDADE'])['CODIGO_CURSO'].count().reset_index()

display(contagem_cursos_por_cidade)
```

8 - Verificando quantidade de cursos oferecidos por Instituição e total de Instituições.

```python
contagem_instituicoes = tabela['NOME_IES'].value_counts().reset_index()

contagem_instituicoes.columns = ['Nome da Instituição', 'Quantidade de Cursos']

total_instituicoes = tabela['NOME_IES'].nunique()

print("Total de Instituições:", total_instituicoes)

display(contagem_instituicoes)
```

9 - Comparativo entre Categoria administrativa com os tipos de organização academica.

```python
tabela_comparativa_org = pd.crosstab(tabela['ORGANIZACAO_ACADEMICA'], tabela['CATEGORIA_ADMINISTRATIVA'])

display(tabela_comparativa_org)
```

10 - Exibindo a quantidade de vagas e a situação do curso por Instituição.

```python
atividade_cidade = tabela.groupby(['NOME_CURSO', 'MUNICIPIO', 'SITUACAO_CURSO'])['QT_VAGAS_AUTORIZADAS'].count().reset_index()

display(atividade_cidade)
```

11 - Verificando a quantidade de Instituições por região.

```python
tabela_comparativa_regiao = tabela.groupby('REGIAO')['NOME_CURSO'].nunique().reset_index()

tabela_comparativa_regiao.columns = ['Região', 'Quantidade de Instituições']

print(tabela_comparativa_regiao)
```

12 - Exibindo a caraga horaria por curso.

```python
tabela_carga_horaria = tabela.groupby('NOME_CURSO')['CARGA_HORARIA'].count().reset_index()

display(tabela_carga_horaria)
```

13 - Exibindo graficos dos dados obtidos para elaboração de documento.

```python
import matplotlib.pyplot as plt


regioes = tabela_comparativa_regiao['Região']
quantidade_instituicoes = tabela_comparativa_regiao['Quantidade de Instituições']

plt.figure(figsize=(10, 6))
plt.barh(regioes, quantidade_instituicoes)
plt.xlabel('Quantidade de Instituições')
plt.ylabel('Região')
plt.title('Quantidade de Instituições por Região')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

total_instituicoes = tabela_comparativa_org.index
data = tabela_comparativa_org.sum(axis=1)

explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
colors = ("orange", "cyan", "brown", "grey", "indigo", "beige")
wp = {'linewidth': 1, 'edgecolor': "green"}

def func(pct, allvalues):
    absolute = int(pct / 100. * np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)

fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), explode=explode, labels=total_instituicoes, shadow=True, colors=colors, startangle=90, wedgeprops=wp, textprops=dict(color="magenta"))

ax.legend(wedges, total_instituicoes, title="Organizações Acadêmicas", loc="center left", bbox_to_anchor=(1.2, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Comparação entre Organizações Acadêmicas e Categorias Administrativas")
plt.show()
```
